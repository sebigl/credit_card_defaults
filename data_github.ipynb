```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

def create_model():
    model = Sequential()
    #21 feature inputs going into an 50-unit layer 
    model.add(Dense(25, input_dim=21, kernel_initializer='normal', activation='relu'))
    #Hidden Layer 1
    model.add(Dense(50, kernel_initializer='normal', activation='relu'))
    #HL 2
    model.add(Dense(30, kernel_initializer='normal', activation='relu'))
    #HL 3
    model.add(Dense(10, kernel_initializer='normal', activation='relu'))
    #HL 4
    #model.add(Dense(5, kernel_initializer='normal', activation='relu'))
    #HL 5
    #model.add(Dense(3, kernel_initializer='normal', activation='relu'))
    #HL 6
    #model.add(Dense(3, kernel_initializer='normal', activation='sigmoid'))
    
    
    # Output layer with a binary classification (default or no default)
    model.add(Dense(1, kernel_initializer='normal', activation='relu'))
    # Compile model; rmsprop seemed to work best
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```
